```python
df['Data de Adesão'] = pd.to_datetime(df['Data de Adesão'], format='%Y-%m-%d %H:%M:%S')
df['Data de Carência'] = pd.to_datetime(df['Data de Carência'], format='%Y-%m-%d %H:%M:%S')
df['Data de Cancelamento'] = pd.to_datetime(df['Data de Cancelamento'], format='%Y-%m-%d %H:%M:%S')
df.head()
```

```python
df_compr = df[(df['Data de Adesão'] >= '2022-01-01')]
df_compr.head()
```

```python
df_compr['mes'] = pd.DatetimeIndex(df_compr['Data de Adesão']).month
# df['mes_canc'] = df['mes_canc'].astype(int, errors='ignore')
df_compr['mes'] = df_compr['mes'].fillna(0).astype(int)

df_compr.head()
```

```python
import calendar
result = []
for i in df_compr['mes']:
    result.append(calendar.month_abbr[i])
df_compr['month'] = result
print(result)
```

```python
df_compr.drop('mes', axis=1, inplace=True)
df_compr
```

```python
df_compr.sort_values("month")
```

# Preparando Ativos

```python
df_compr2 = df_compr[['mes']
df_compr2
```

```python
import calendar
result = []
for i in df_compr2['mes']:
    result.append(calendar.month_abbr[i])
df_compr2['month'] = result
print(result)
```

```python
df_compr2
```

```python
df_compr_gr2 = (df_compr2.groupby(['month'])['Preço'].agg(['sum']).sort_values(by='sum', ascending=False).reset_index())
df_compr_gr2
```

```python
months = ["Jan", 
          "Feb", "Mar", "Apr"]
df_compr_gr2['month'] = pd.Categorical(df_compr_gr2['month'], categories=months, ordered=True)
df_compr_gr2.sort_values(by='month', ascending = True, inplace=True)

df_compr_gr2
```

```python
df_compr_gr2.sort_values("month")
```

```python
df_compr_gr2.rename(columns = {'sum':'Total'}, inplace = True)
df_compr_gr2
```

```python
df_compr_gr2.to_csv("../datasets/sales_mensal.csv")
```
